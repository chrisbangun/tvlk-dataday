```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.rcParams['figure.figsize'] = (8, 8)
```

```python
def plot_data(data, labels):
    
    positives = data[labels == 1, :]
    negatives = data[labels == 0, :]
    
    plt.scatter(positives[:, 0], positives[:, 1], color='red', marker='+', s=200)
    plt.scatter(negatives[:, 0], negatives[:, 1], color='blue', marker='_', s=200)

def plot_boundary():
    pass
```

```python
positives = np.array([[1, 0], [0, 1]])
negatives = np.array([[0, 0], [1, 1]])

data = np.concatenate([positives, negatives])
labels = np.array([1, 1, 0, 0])
```

```python
plot_data(data, labels)
```

```python
w1 = -0.5
w2 = 0.5
threshold = 0.1

resolution = 100
xs = np.linspace(-0.1, 1.1, resolution)
scores = np.zeros((resolution, resolution))

for i, x1 in enumerate(xs):
    for j, x2 in enumerate(xs):
        scores[i, j] = w1*x1 + w2*x2 > threshold
```

```python
plt.contour(xs, xs, scores, alpha=0.5, colors='black')
plt.contourf(xs, xs, scores, colors=('#ff0000', '#0000ff', '#0000ff'), levels=(-1, 0, 1), alpha=0.2)

plot_data(data, labels)
```

```python
# two layer network
w11 = 1
w12 = 1
threshold1 = 0.5

w21 = -1
w22 = -1
threshold2 = -1.5

w31 = 1
w32 = 1
threshold3 = 1
```

```python
resolution = 100
xs = np.linspace(-0.1, 1.1, resolution)

y1 = np.zeros((resolution, resolution))
y2 = np.zeros((resolution, resolution))
y3 = np.zeros((resolution, resolution))

for i, x1 in enumerate(xs):
    for j, x2 in enumerate(xs):
        
        y1[i, j] = w11*x1 + w12*x2 > threshold1
        y2[i, j] = w21*x1 + w22*x2 > threshold2
        y3[i, j] = w31*y1[i, j] + w32*y2[i, j] > threshold3
```

```python
plt.contour(xs, xs, y1, alpha=0.5, colors='black')
plt.contourf(xs, xs, y1, colors=('#0000ff', '#ff0000', '#00ff00'), levels=(-1, 0, 1), alpha=0.2)

plot_data(data,labels)
```

```python
plt.contour(xs, xs, y2, alpha=0.5, colors='black')
plt.contourf(xs, xs, y2, colors=('#0000ff', '#ff0000', '#00ff00'), levels=(-1, 0, 1), alpha=0.2)

plot_data(data, labels)
```

```python
plt.contour(xs, xs, y3, alpha=0.5, colors='black')
plt.contourf(xs, xs, y3, colors=('#0000ff', '#ff0000', '#00ff00'), levels=(-1, 0, 1), alpha=0.2)

plt.scatter(positives[:, 0], positives[:, 1], color='red', marker='+', s=200)
plt.scatter(negatives[:, 0], negatives[:, 1], color='blue', marker='_', s=200)
```
